1. Tentukan luas jajaran genjang yang ditentukan oleh vektor

$$
\vec{u} = \begin{bmatrix}1 \\ 2\end{bmatrix}, \quad \vec{v} = \begin{bmatrix}2 \\ 1\end{bmatrix}
$$

#### **Penyelesaian:**

Gunakan determinan 2x2:

$$
\text{Luas} = \left| \begin{vmatrix}1 & 2 \\ 2 & 1\end{vmatrix} \right| = |(1)(1) - (2)(2)| = |1 - 4| = |-3| = 3
$$

---

2. Tentukan luas jajaran genjang yang ditentukan oleh vektor

$$
\vec{u} = \begin{bmatrix}2 \\ 0\end{bmatrix}, \quad \vec{v} = \begin{bmatrix}0 \\ 3\end{bmatrix}
$$

#### **Penyelesaian:**

$$
\text{Luas} = \left| \begin{vmatrix}2 & 0 \\ 0 & 3\end{vmatrix} \right| = |(2)(3) - (0)(0)| = |6| = 6
$$

---

3. Tentukan luas segitiga dengan titik-titik sudut:

$$
(0, 0, 0),\quad (1, 3, -1),\quad (2, 1, 1)
$$

#### **Penyelesaian:**

1. Hitung dua vektor:

   $$
   \vec{AB} = (1, 3, -1),\quad \vec{AC} = (2, 1, 1)
   $$
2. Hitung cross product:

   $$
   \vec{AB} \times \vec{AC} = \begin{vmatrix}
   \hat{i} & \hat{j} & \hat{k} \\
   1 & 3 & -1 \\
   2 & 1 & 1
   \end{vmatrix}
   = (4, -3, -5)
   $$
3. Panjang vektor:

   $$
   |\vec{n}| = \sqrt{4^2 + (-3)^2 + (-5)^2} = \sqrt{50}
   $$
4. Luas segitiga:

   $$
   \frac{1}{2} \cdot \sqrt{50} = \frac{5\sqrt{2}}{1} \approx 3.54
   $$

4. Tentukan luas segitiga dengan titik-titik sudut:

$$
(5, 2, -1),\quad (3, 6, 2),\quad (1, 0, 4)
$$

#### **Penyelesaian:**

1. Hitung vektor:

   $$
   \vec{AB} = (-2, 4, 3),\quad \vec{AC} = (-4, -2, 5)
   $$
2. Cross product:

   $$
   \vec{AB} \times \vec{AC} = (26, -2, 20)
   $$
3. Panjang vektor:

   $$
   |\vec{n}| = \sqrt{26^2 + (-2)^2 + 20^2} = \sqrt{1080}
   $$
4. Luas segitiga:

   $$
   \frac{1}{2} \cdot \sqrt{1080} \approx 16.43
   $$
