# Perkalian Silang (Cross Product)

## Definisi Hasil Kali Silang (Cross Product)

Hasil kali silang (cross product) adalah operasi antara dua vektor dalam ruang tiga dimensi (3D) yang menghasilkan **vektor baru**. Vektor hasil ini:

* Tegak lurus terhadap kedua vektor asal (mengikuti aturan tangan kanan).
* Memiliki arah dan panjang (magnitudo).

Rumus umum:
Jika dua vektor diberikan sebagai:

$$
\vec{A} = \langle A_x, A_y, A_z \rangle,\quad \vec{B} = \langle B_x, B_y, B_z \rangle
$$

maka:

$$
\vec{A} \times \vec{B} = \langle A_yB_z - A_zB_y,\; A_zB_x - A_xB_z,\; A_xB_y - A_yB_x \rangle
$$

---

## Contoh Menghitung Cross Product

### Diketahui:

$$
\vec{A} = \langle 2, 3, 4 \rangle,\quad \vec{B} = \langle 5, 6, 7 \rangle
$$

### Langkah-langkah Penyelesaian:

1. **Gunakan rumus komponen:**

$$
\vec{A} \times \vec{B} = \langle A_yB_z - A_zB_y,\; A_zB_x - A_xB_z,\; A_xB_y - A_yB_x \rangle
$$

2. **Hitung masing-masing komponen:**

   * Komponen $x$:

     $$
     A_yB_z - A_zB_y = 3 \cdot 7 - 4 \cdot 6 = 21 - 24 = -3
     $$

   * Komponen $y$:

     $$
     A_zB_x - A_xB_z = 4 \cdot 5 - 2 \cdot 7 = 20 - 14 = 6
     $$

   * Komponen $z$:

     $$
     A_xB_y - A_yB_x = 2 \cdot 6 - 3 \cdot 5 = 12 - 15 = -3
     $$

3. **Gabungkan hasilnya:**

$$
\vec{A} \times \vec{B} = \langle -3, 6, -3 \rangle
$$

---

## Menggunakan Determinan untuk Menemukan Cross Product


Cross product bisa dihitung dengan **determinan dari matriks 3x3** yang disusun seperti ini:

$$
\vec{A} \times \vec{B} =
\begin{vmatrix}
\hat{i} & \hat{j} & \hat{k} \\
A_x & A_y & A_z \\
B_x & B_y & B_z \\
\end{vmatrix}
$$

### Contoh

$$
\vec{A} = \langle 1, 2, 3 \rangle,\quad \vec{B} = \langle 4, 5, 6 \rangle
$$

### Langkah-langkah Penyelesaian:

1. **Bentuk matriks determinan:**

$$
\vec{A} \times \vec{B} =
\begin{vmatrix}
\hat{i} & \hat{j} & \hat{k} \\
1 & 2 & 3 \\
4 & 5 & 6 \\
\end{vmatrix}
$$

2. **Hitung komponen $\hat{i}$:**
   Tutup kolom $\hat{i}$, sisanya:

   $$
   \begin{vmatrix}
   2 & 3 \\
   5 & 6 \\
   \end{vmatrix}
   = 2 \cdot 6 - 3 \cdot 5 = 12 - 15 = -3
   $$

3. **Hitung komponen $\hat{j}$ (diberi tanda negatif):**
   Tutup kolom $\hat{j}$, sisanya:

   $$
   \begin{vmatrix}
   1 & 3 \\
   4 & 6 \\
   \end{vmatrix}
   = 1 \cdot 6 - 3 \cdot 4 = 6 - 12 = -6 \Rightarrow +6 (\text{karena tanda -})
   $$

4. **Hitung komponen $\hat{k}$:**
   Tutup kolom $\hat{k}$, sisanya:

   $$
   \begin{vmatrix}
   1 & 2 \\
   4 & 5 \\
   \end{vmatrix}
   = 1 \cdot 5 - 2 \cdot 4 = 5 - 8 = -3
   $$

5. **Gabungkan hasilnya:**

$$
\vec{A} \times \vec{B} = -3\hat{i} + 6\hat{j} - 3\hat{k} = \langle -3, 6, -3 \rangle
$$


```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Definisikan vektor A dan B
A = np.array([1, 2, 3])
B = np.array([4, 5, 6])

# Hitung cross product dengan determinan
i = A[1]*B[2] - A[2]*B[1]
j = -(A[0]*B[2] - A[2]*B[0])
k = A[0]*B[1] - A[1]*B[0]
C = np.array([i, j, k])  # Cross Product A × B

# Cetak hasil
print("Vektor A:", A)
print("Vektor B:", B)
print("Cross Product A × B:", C)
```

## Sifat-Sifat Perkalian Silang (Cross Product)

Berikut beberapa sifat dari perkalian silang (Cross Product):

### 1. Antikomutatif:

$\vec{A} \times \vec{B} = - (\vec{B} \times \vec{A})$
Contoh:
$\vec{A} = \langle 1, 0, 0 \rangle, \vec{B} = \langle 0, 1, 0 \rangle$
$\vec{A} \times \vec{B} = \langle 0, 0, 1 \rangle$
$\vec{B} \times \vec{A} = \langle 0, 0, -1 \rangle$

### 2. Jika vektor sejajar, hasilnya nol:

$\vec{A} \parallel \vec{B} \Rightarrow \vec{A} \times \vec{B} = \vec{0}$
Contoh:
$\vec{A} = \langle 2, 4, 6 \rangle, \vec{B} = \langle 1, 2, 3 \rangle \Rightarrow \vec{A} = 2\vec{B} \Rightarrow \vec{A} \times \vec{B} = \vec{0}$

### 3. Hasil $\vec{A} \times \vec{B}$ tegak lurus terhadap A dan B:

Contoh:
$\vec{A} = \langle 1, 0, 0 \rangle, \vec{B} = \langle 0, 1, 0 \rangle \Rightarrow \vec{A} \times \vec{B} = \langle 0, 0, 1 \rangle$

### 4. Distribusi terhadap penjumlahan:

$\vec{A} \times (\vec{B} + \vec{C}) = \vec{A} \times \vec{B} + \vec{A} \times \vec{C}$
Contoh:
$\vec{A} = \langle 1, 0, 0 \rangle, \vec{B} = \langle 0, 1, 0 \rangle, \vec{C} = \langle 0, 0, 1 \rangle$
$\vec{A} \times (\vec{B} + \vec{C}) = \langle 0, -1, 1 \rangle$

### 5. Perkalian dengan skalar:

$k(\vec{A} \times \vec{B}) = (k\vec{A}) \times \vec{B} = \vec{A} \times (k\vec{B})$
Contoh:
$k = 3, \vec{A} = \langle 1, 0, 0 \rangle, \vec{B} = \langle 0, 1, 0 \rangle$
$k(\vec{A} \times \vec{B}) = \langle 0, 0, 3 \rangle$

### 6. Tidak asosiatif:

$(\vec{A} \times \vec{B}) \times \vec{C} \neq \vec{A} \times (\vec{B} \times \vec{C})$
Contoh:
$\vec{A} = \langle 1, 0, 0 \rangle, \vec{B} = \langle 0, 1, 0 \rangle, \vec{C} = \langle 0, 0, 1 \rangle$

### 7. Vektor dikali dirinya sendiri = nol:

$\vec{A} \times \vec{A} = \vec{0}$
Contoh:
$\vec{A} = \langle 2, -1, 5 \rangle \Rightarrow \vec{A} \times \vec{A} = \vec{0}$

## 3.2 Hasil Kali Silang dan Sudut Antar Vektor

Jika $\vec{A}$ dan $\vec{B}$ adalah dua vektor dengan sudut $\theta$ di antara mereka, maka besar hasil kali silang diberikan oleh:

$|\vec{A} \times \vec{B}| = |\vec{A}| \cdot |\vec{B}| \cdot \sin\theta$

### Contoh:

Diberikan dua vektor:
$\vec{A} = \langle 3, 0, 0 \rangle, \quad \vec{B} = \langle 0, 4, 0 \rangle$
Hitung besar cross product dan sudut antara vektor.

**Langkah-langkah:**

1. Hitung panjang vektor:
   $|\vec{A}| = \sqrt{3^2 + 0^2 + 0^2} = 3$
   $|\vec{B}| = \sqrt{0^2 + 4^2 + 0^2} = 4$
2. Hitung hasil cross product:
   $\vec{A} \times \vec{B} = \langle 0, 0, 12 \rangle \Rightarrow |\vec{A} \times \vec{B}| = 12$
3. Gunakan rumus:
   $|\vec{A} \times \vec{B}| = |\vec{A}| \cdot |\vec{B}| \cdot \sin\theta \Rightarrow 12 = 3 \cdot 4 \cdot \sin\theta$
4. Selesaikan untuk $\theta$:
   $\sin\theta = \frac{12}{12} = 1 \Rightarrow \theta = 90^\circ$
---

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import math

# Definisikan vektor A dan B
A = np.array([3, 0, 0])
B = np.array([0, 4, 0])

# Hitung panjang masing-masing vektor
magnitude_A = np.linalg.norm(A)
magnitude_B = np.linalg.norm(B)

# Hitung hasil cross product
cross = np.cross(A, B)
magnitude_cross = np.linalg.norm(cross)

# Gunakan rumus: |A × B| = |A||B|sinθ → sinθ = |A × B| / (|A||B|)
sin_theta = magnitude_cross / (magnitude_A * magnitude_B)
theta_radian = np.arcsin(sin_theta)
theta_degree = np.degrees(theta_radian)

# Cetak hasil
print("Cross Product A × B =", cross)
print("Magnitude Cross Product =", magnitude_cross)
print("Sudut antara A dan B =", theta_degree, "derajat")

# Visualisasi 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Gambar vektor A, B, dan cross product
ax.quiver(0, 0, 0, A[0], A[1], A[2], color='r', label='A', arrow_length_ratio=0.1)
ax.quiver(0, 0, 0, B[0], B[1], B[2], color='g', label='B', arrow_length_ratio=0.1)
ax.quiver(0, 0, 0, cross[0], cross[1], cross[2], color='b', label='A × B', arrow_length_ratio=0.1)

# Atur tampilan
ax.set_xlim([0, 4])
ax.set_ylim([0, 4])
ax.set_zlim([0, 13])
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title(f'Sudut antara A dan B = {theta_degree:.1f}°')
ax.legend()
plt.show()
```

## Implementasi Cross Product dalam Luas Segitiga

Cross product dapat digunakan untuk menghitung **luas segitiga** yang dibentuk oleh dua vektor:

$\text{Luas} = \frac{1}{2} |\vec{A} \times \vec{B}|$

### Contoh Soal:

Titik-titik: $P(0, 0, 0), Q(2, 3, 0), R(4, 1, 0)$
Hitung luas segitiga PQR.

### Langkah-langkah:

1. Buat vektor:
   $\vec{PQ} = \langle 2, 3, 0 \rangle$, $\vec{PR} = \langle 4, 1, 0 \rangle$
2. Hitung cross product:

$$
\vec{PQ} \times \vec{PR} =
\begin{vmatrix}
\hat{i} & \hat{j} & \hat{k} \\
2 & 3 & 0 \\
4 & 1 & 0
\end{vmatrix}
= \langle 0, 0, -10 \rangle
$$

3. Magnitudo:
   $|\vec{PQ} \times \vec{PR}| = 10$
4. Luas:
   $\frac{1}{2} \cdot 10 = 5$

### Jawaban:

Luas segitiga = **5 satuan luas**



### Visualisasi dengan Python (PCA dan Plotting)

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.decomposition import PCA

# Titik-titik segitiga
P = np.array([0, 0, 0])
Q = np.array([2, 3, 0])
R = np.array([4, 1, 0])

# Vektor dari P
PQ = Q - P
PR = R - P

# Cross product dan luas
cross = np.cross(PQ, PR)
area = 0.5 * np.linalg.norm(cross)

# Proyeksi 2D dengan PCA
X = np.vstack([P, Q, R])
pca = PCA(n_components=2)
X_proj = pca.fit_transform(X)

# Plot hasil
plt.figure(figsize=(6,6))
plt.fill(X_proj[:,0], X_proj[:,1], 'skyblue', edgecolor='black', alpha=0.5)
plt.scatter(X_proj[:,0], X_proj[:,1], c='red')
for i, label in enumerate(['P', 'Q', 'R']):
    plt.text(X_proj[i,0], X_proj[i,1], label)
plt.title(f'Luas Segitiga = {area}')
plt.axis('equal')
plt.grid(True)
plt.show()
```

```python
# Titik-titik segitiga
P = (0, 0)
Q = (2, 3)
R = (4, 1)

# Fungsi untuk membuat vektor dari dua titik
def make_vector(p1, p2):
    return (p2[0] - p1[0], p2[1] - p1[1])

# Fungsi untuk menghitung cross product 2D
def cross_product(v1, v2):
    return v1[0] * v2[1] - v1[1] * v2[0]

# Buat dua vektor dari titik P
PQ = make_vector(P, Q)
PR = make_vector(P, R)

# Hitung cross product
cp = cross_product(PQ, PR)

# Hitung luas segitiga (setengah dari magnitude cross product)
area = 0.5 * abs(cp)

# Cetak hasil
print("Vektor PQ:", PQ)
print("Vektor PR:", PR)
print("Cross Product PQ × PR:", cp)
print("Luas Segitiga:", area)
```
